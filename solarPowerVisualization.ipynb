```python
import matplotlib.pyplot as plt
```

```python
# plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot()
```

```python
plt.Figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('distribution of generated power (kw)')
plt.xlabel('Generated power (kw)')
plt.ylabel('frequency')
plt.show()
```

```python
# Histogram: Distribution of temperature with a different style
plt.figure(figsize=(10, 6))
sns.histplot(df['temperature_2_m_above_gnd'], bins=25, kde=True, color='teal', edgecolor="black")
plt.title('Temperature Distribution (2m above ground)', fontsize=16, fontweight='bold')
plt.xlabel('Temperature (Â°C)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Scatter Plot: Generated power vs wind speed with color intensity based on humidity
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['wind_speed_10_m_above_gnd'], df['generated_power_kw'], 
                       c=df['relative_humidity_2_m_above_gnd'], cmap='viridis', alpha=0.7, edgecolors='k')
plt.colorbar(scatter, label='Relative Humidity (%)')
plt.title('Generated Power vs Wind Speed with Humidity Intensity', fontsize=16)
plt.xlabel('Wind Speed (10m above ground) (m/s)', fontsize=12)
plt.ylabel('Generated Power (kW)', fontsize=12)
plt.show()
```

```python
# Area Plot: Cumulative precipitation over time
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values(by='temperature_2_m_above_gnd').reset_index()
cumulative_precipitation = df_sorted['total_precipitation_sfc'].cumsum()
plt.fill_between(df_sorted.index, cumulative_precipitation, color='lightblue', alpha=0.5)
plt.plot(df_sorted.index, cumulative_precipitation, color='blue', lw=2)
plt.title('Cumulative Precipitation Over Sorted Temperature', fontsize=16)
plt.xlabel('Index (Sorted by Temperature)', fontsize=12)
plt.ylabel('Cumulative Precipitation (mm)', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
